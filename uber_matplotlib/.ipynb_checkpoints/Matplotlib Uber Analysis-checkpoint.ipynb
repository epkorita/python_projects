```python
%matplotlib inline

# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

city_file = './Resources/city_data.csv'
ride_file = './Resources/ride_data.csv'

# Read the City and Ride Data
city_data = pd.read_csv(city_file)
ride_data = pd.read_csv(ride_file)

# Combine the data into a single dataset
complete_data = pd.merge(ride_data, city_data, how = "left", on = ["city", "city"])

# Display the data table for preview
complete_data.head()
```

```python
x = complete_data.groupby(['type','city'])['ride_id'].count()
y = complete_data.groupby(['type','city'])['fare'].mean()
z = complete_data.groupby(['type','city'])['driver_count'].mean()

colors = {'Urban':'coral', 'Suburban':'lightskyblue', 'Rural':'gold'}
pie_colors = 'gold','lightskyblue','lightcoral'
recs = [colors[r] for r in x.reset_index()['type']]

urban_l = plt.Circle([], [], color='coral', label='Urban')
suburban_l = plt.Circle([], [], color='lightskyblue', label='Suburban')
rural_1 = plt.Circle([], [], color='gold', label='Rural')

handles = [urban_l,suburban_l,rural_1]
labels = [h.get_label() for h in handles]

# plotting 

plt.scatter(x, y,s=z*10,marker="o",edgecolors="black",alpha=.80,c=recs)
plt.title('PyBer Ride Sharing Data(2016)', fontsize=20)
plt.xlabel('Total Number of Rides(Per City)', fontsize=12)
plt.ylabel('Average Fare($)', fontsize=12)
plt.legend(handles=handles,labels=labels,title='City Types')
plt.annotate('    Note:', xy=(40, 37))
plt.annotate('    Circle size correlates with driver count per city', xy=(40, 35.5))
plt.grid()
# plt.savefig('ride_sharing_data.png')
plt.show()
```

```python
percentage_fare = 100*complete_data.groupby('type')['fare'].sum()/complete_data['fare'].sum()
explode = (0, 0, 0.1)
labels = 'Rural', 'Suburban', 'Urban'
plt.pie(percentage_fare,autopct='%.2f%%',explode=explode,labels=labels,shadow=True,colors=pie_colors)
plt.title('% of Total Fares by City Type')
# plt.savefig('percent_fares.png')
plt.show()
```

```python
percentage_rides = 100*complete_data.groupby('type')['ride_id'].count()/complete_data['ride_id'].count()
explode = (0, 0, 0.1)
labels = 'Rural', 'Suburban', 'Urban'
plt.pie(percentage_rides,autopct='%.2f%%',explode=explode,labels=labels,shadow=True, colors=pie_colors)
plt.title('% of Total Rides by City Type')
# plt.savefig('percent_rides.png')
plt.show()
```

```python
percentage_drivers = 100*city_data.groupby('type')['driver_count'].sum()/city_data['driver_count'].sum()
explode = (0, 0, 0.1)
labels = 'Rural', 'Suburban', 'Urban'
plt.pie(percentage_drivers,autopct='%.2f%%',explode=explode,shadow=True,labels=labels,colors=pie_colors)
plt.title('% of Total Drivers by City Type')
# plt.savefig('percent_drivers.png')
plt.show()
```
